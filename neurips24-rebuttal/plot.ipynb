```python
import matplotlib.pyplot as plt
import numpy as np

# Data
n = np.array([200, 500, 1000, 2000])
u_delta_f = np.array([A_200[-1], A_500[-1], A_1000[-1], A_2000[-1]])
u_f = np.array([reg_200[-1], reg_500[-1], reg_1000[-1], reg_2000[-1]])

# Plot
plt.figure(figsize=(8, 6))
plt.loglog(n, u_delta_f, 'o-', label='$u = \Delta f$')
plt.loglog(n, u_f, 's-', label='$u = f$')

# Labels and title
plt.xlabel('$n$', fontsize=14)
plt.ylabel('Relative $L^2$ error', fontsize=14)
plt.title('Log-Log Plot of Relative $L^2$ Error', fontsize=16)
plt.legend()

# Set ticks for log scale
plt.xticks([200, 500, 1000, 2000], ['200', '500', '1000', '2000'])
plt.yticks([1e-3, 1e-2], ['$10^{-3}$', '$10^{-2}$'])

plt.grid(True, which="both", ls="--")

# Show plot
plt.show()
```
