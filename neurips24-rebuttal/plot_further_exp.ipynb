```python
# log log plot
# x_axis: 200, 500, 1000, 2000
# plot relu, relu2, relu3

# import file
import numpy as np
import matplotlib.pyplot as plt

# read data
# load numpy array

# log log plot
plt.figure()
plt.xlabel('sample size')
plt.ylabel('relative L^2 error')

# Set ticks for log scale
plt.xticks([50, 100, 200, 500, 1000, 2000], ['50', '100', '200', '500', '1000', '2000'])
plt.yticks([1e-3, 1e-2], ['$10^{-3}$', '$10^{-2}$'])

n = [50, 100, 200, 500, 1000, 2000]
for act in ['relu', 'relu2', 'relu3']:
    content = []
    for sample_size in n:
        path = f'/home/howon/neurips24-rebuttal/Reproducing-Activation-Function/ScientificComputing/regular/further_exp/error_save_mlp_{act}_La_2D_w256_bs_{sample_size}_poly_sin_gaussian.npy'
        data = np.load(path)
        content.append(data[-1])
        #print(data[-1])

        # plot
    plt.loglog(n, content, 'o-', label=act)
        #plt.plot(data[-1], label=act)
plt.title('The Scaling law for width = 256')
plt.legend()
plt.xticks([50, 100, 200, 500, 1000, 2000], ['50', '100', '200', '500', '1000', '2000'])
plt.yticks([1e-3, 1e-2], ['$10^{-3}$', '$10^{-2}$'])
plt.show()
```
